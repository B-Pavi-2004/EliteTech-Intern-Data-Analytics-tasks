# Data cleaning

```python

df = df.dropna()
```

```python

df = df.drop_duplicates()
```

# Exploratory Data Analysis

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Distribution of Carat

sns.histplot(df['carat'], kde=True)
plt.title('Distribution of Carat')
plt.show()


# Distribution of Price

sns.histplot(df['price'], kde=True)
plt.title('Distribution of Price')
plt.show()
```

```python
# Scatter plot of Carat vs Price


sns.scatterplot(x='carat', y='price', data=df)
plt.title('Carat vs Price')
plt.show()
```

```python
# Box plot of Cut vs Price


sns.boxplot(x='cut', y='price', data=df)
plt.title('Cut vs Price')
plt.show()
```

```python
# Group by Cut and find average price

avg_price_by_cut = df.groupby('cut')['price'].mean()
print(avg_price_by_cut)

avg_price_by_cut.plot(kind='bar', title='Average Price by Cut')
plt.show()
```

```python
# Violin plot of price across different cuts
sns.violinplot(x='cut', y='price', data=df)
plt.title('Violin Plot of Price by Cut')
plt.show()
```

```python
# Joint plot of Carat vs Price
sns.jointplot(x='carat', y='price', data=df, kind='hex', gridsize=30, cmap='Blues')
plt.show()
```

```python
# Pivot table for Cut, Color, and average Price
pivot = df.pivot_table(values='price', index='cut', columns='color', aggfunc='mean')

# Heatmap of Cut and Color vs Average Price
sns.heatmap(pivot, annot=True, cmap='coolwarm')
plt.title('Heatmap of Cut and Color vs Average Price')
plt.show()
```

```python
# Box plot for Carat vs Price across different Cuts and Colors
plt.figure(figsize=(10, 6))
sns.boxplot(x='cut', y='price', hue='color', data=df)
plt.title('Box Plot of Price by Cut and Color')
plt.show()
```

```python
# Line plot (example)
sns.lineplot(x='carat', y='price', data=df)
plt.title('Line Plot of Carat vs Price')
plt.show()
```

```python
# Count plot for Cut categories
sns.countplot(x='cut', data=df)
plt.title('Count of Diamonds by Cut')
plt.show()
```

```python
import numpy as np
sns.histplot(np.log(df['price']), kde=True)
plt.title('Log-transformed Price Distribution')
plt.show()
```
